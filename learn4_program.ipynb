异常处理机制

```python
#异常现象（ExceptionObject)
#回溯traceback
#try except、多个except的try语句、try except finally语句结构进行异常处理
#try:
#    <语句>#运行别的代码
#except <name>:
#    <语句>#如果发生了try部分异常，那么执行这个语句
#except <name>,<data>:
#    <语句>#如果发生了name异常，那么执行这个语句，可以使用data获取异常信息
#else:
#    <语句>#如果没有异常发生
#finally:
#    <语句>#无论是否发生异常都会执行

#异常处理
#捕捉单个异常
try:
    print(8/0)
except ZeroDivisionError as err:
    print("发生异常",err)
#捕捉多个异常
try:
    print(8/0)
except(ZeroDivisionError,TypeError):
    print("发生了一个异常") 
#捕捉所有异常
try:
    print(8/0)
except:
    print("程序出现异常")


x=[1,2]
try:
    try:
        8/0
    except ZeroDivisionError:
        print("内部除零异常")
        x[2]/2
except IndexError:
    print("索引异常")

#异常处理结构的嵌套
with open("filename","w")as file:
    try:
        file.write("hello world")
    except IOError:
        print("文件写入异常")
    finally:
        print("文件关闭")
#异常处理结构的嵌套
#with open("test.txt","w")as outfile:
#    write output to outfile.


try:
    file = open("example.txt", "r")
    content = file.read()
    print(content)
except FileNotFoundError as e:
    print("文件未找到:", e)
except IOError as e:
    print("IO错误:", e)
else:
    print("文件读取成功")
finally:
    if 'file' in locals():
        file.close()
        print("文件已关闭")
```
